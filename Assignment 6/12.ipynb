# SVM - Email Spam Classifier
#### Libraries Used

```python
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from IPython.display import clear_output
from sklearn.svm import SVC
import pandas as pd
```

#### Input File path

```python
datasetPath = "spambase.data"
```

#### Headers and reading input data into pandas dataframe

```python
headers = ["word_freq_make", "word_freq_address", "word_freq_all", "word_freq_3d", "word_freq_our", 
           "word_freq_over", "word_freq_remove", "word_freq_internet", "word_freq_order ", 
           "word_freq_mail", "word_freq_receive", "word_freq_will", "word_freq_people", 
           "word_freq_report", "word_freq_addresses", "word_freq_free", "word_freq_business", 
           "word_freq_email", "word_freq_you", "word_freq_credit", "word_freq_your", "word_freq_font", 
           "word_freq_000", "word_freq_money", "word_freq_hp", "word_freq_hpl", "word_freq_george", 
           "word_freq_650", "word_freq_lab", "word_freq_labs", "word_freq_telnet", "word_freq_857", 
           "word_freq_data", "word_freq_415", "word_freq_85", "word_freq_technology", "word_freq_1999",
           "word_freq_parts", "word_freq_pm", "word_freq_direct", "word_freq_cs", "word_freq_meeting", 
           "word_freq_original", "word_freq_project", "word_freq_re", "word_freq_edu", "word_freq_table", 
           "word_freq_conference", "char_freq_;", "char_freq_(", "char_freq_[", "char_freq_!", "char_freq_$", 
           "char_freq_#", "capital_run_length_average", "capital_run_length_longest", "capital_run_length_total", "class"]

df = pd.read_csv(datasetPath, names = headers)
df
```

#### Seperating the label (class) column for training

```python
data = df.drop(["class"], axis='columns')  #Stores the data except the last label (class)
label = df.iloc[:,-1]                      #Stores the label (class) column
```

#### Pre-processing data with StandardScaler for better training and 
#### Splitting the input data into 70% training and 30% testing sets

```python
data = StandardScaler().fit_transform(data)
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size = 0.3, random_state = 3)
print("Y_Train mean:", y_train.mean())
print("Y_Test mean", y_test.mean())
```

#### Set of all C-Values and Kernals that are used

```python
CValues = [0.0005, 0.0001, 0.005, 0.001, 0.05, 0.01, 0.5, 0.1, 1, 5, 10, 50, 100, 500, 1000, 5000]
kernals = ["linear", "poly", "rbf"]
```

#### Training, testing and printing results

```python
result = pd.DataFrame(columns=["C_Value", "Linear_Train", "Linear_Test", "Poly_Train", "Poly_Test", "RBF_Train", "RBF_Test"])
for cval in CValues:
    accuracies = []
    for ker in kernals:
        #print("Starting")
        #print("C-Value:", cval)
        #print("Kernal:", ker)
        classifier = SVC(C=cval, kernel=ker , max_iter=1e8)
        classifier.fit(x_train, y_train)
        y_pred_train = classifier.predict(x_train)
        y_pred_test = classifier.predict(x_test)
        accuracies.append(accuracy_score(y_train, y_pred_train))
        accuracies.append(accuracy_score(y_test, y_pred_test))
        #print("Done")
        #print()

    result = result.append({'C_Value': cval, "Linear_Train": accuracies[0], "Linear_Test": accuracies[1], 
                          "Poly_Train": accuracies[2], "Poly_Test": accuracies[3],
                          "RBF_Train": accuracies[4], "RBF_Test": accuracies[5]}, ignore_index = True)
    clear_output(wait = True)
    display(result)
```
